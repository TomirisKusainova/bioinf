```python
import pandas as pd
import numpy as np


def RPKM_to_TPM(read_01):
    RPKM_sum = read_01.sum()
    read_01['468_MeCP2_NTC - linear total RPKM'] = read_01['468_MeCP2_NTC - linear total RPKM'].apply(lambda x : x*1000000/RPKM_sum['468_MeCP2_NTC - linear total RPKM'])
    read_01['468_MeCP2_sh1 - linear total RPKM'] = read_01['468_MeCP2_sh1 - linear total RPKM'].apply(lambda x : x*1000000/RPKM_sum['468_MeCP2_sh1 - linear total RPKM'])
    read_01['468_MeCP2_sh3 - linear total RPKM'] = read_01['468_MeCP2_sh3 - linear total RPKM'].apply(lambda x : x*1000000/RPKM_sum['468_MeCP2_sh3 - linear total RPKM'])
    read_01 = read_01.rename(columns = {'468_MeCP2_NTC - linear total RPKM' : "TPM_01",'468_MeCP2_sh1 - linear total RPKM' : "TPM_02",'468_MeCP2_sh3 - linear total RPKM' : "TPM_03"})
    return read_01 

read_01 = pd.read_csv("01. RPKM.tsv",sep="\t", index_col=0)
RPKM_to_TPM(read_01)



```
